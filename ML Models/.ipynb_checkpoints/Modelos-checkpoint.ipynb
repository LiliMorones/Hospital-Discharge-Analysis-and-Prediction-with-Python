## Functions

### Functions to Convert Categorical Variables to Numeric

#### Get dummies for variables with 2 categories

```python
def cat_dummies(df, variables):
    """
    Function that encodes the specified categorical variables in the dataframe into numerical variables, 
    creating a column for each category with values of 1 or 0 (omitting the first category).
    Args:
    df: the dataframe from which the variables to be encoded will be taken.
    variables: list of categorical variables to be converted to numerical.
    Return:
    df: the same dataframe modified with the encoded variables.
    """
    df = pd.get_dummies(df, columns = variables, drop_first=True)
    return df
```

#### Label encoder for nominal variables with more than 2 categories

```python
def cat_label(df, variables):
    """
    Function that encodes the desired nominal categorical variables 
    of a dataframe into numerical variables, where each category is 
    assigned a different number (retaining only one column).
    Args:
    df: the dataframe from which the variables to be encoded will be taken.
    variables: list of categorical variables to be converted to numerical.
    Return:
    df: the same dataframe modified with the encoded variables.
    label_dict: a dictionary that stores the number assigned to each category 
    for each of the encoded variables.
    """
    label_encoder = LabelEncoder()
    diccionario_label = {}
    for v in variables:
        df[v] = label_encoder.fit_transform(df[v])
        diccionario_label[v] = {
            'etiquetas': label_encoder.classes_, 
            'valores_numericos': label_encoder.transform(label_encoder.classes_) 
        }
    return df, diccionario_label
```

```python
def num_etiquetas(variable, diccionario):
    """
    Function that returns the original categories of a variable and their 
    corresponding numeric values after applying LabelEncoder.
    Args:
    variable: the variable for which you want to know its encoding.
    diccionario: a dictionary that contains all the labels from a dataframe.
    Return:
    Direct printing of each label with its respective numeric value.
    """
    etiquetas = diccionario[variable]['etiquetas']
    valores = diccionario[variable]['valores_numericos']
    for etiqueta, valor in zip(etiquetas, valores):
        print(f'{etiqueta}: {valor}')
```

#### Ordinal encoder for ordinal categorical variables

```python
def cat_ordinal(df, categorias):
    """
    Function that encodes the desired ordinal categorical variables
    in a dataframe into numerical variables.
    Args:
    df: the dataframe from which the variables to be encoded will be taken.
    categories: a dictionary with the desired order of categories for each variable to be encoded.
    Return:
    df: the same dataframe with the variables already encoded.
    """
    for columna, cats in categorias.items():
        ordinal_encoder = OrdinalEncoder(categories=[cats]) 
        df[columna] = ordinal_encoder.fit_transform(df[[columna]]) 
    return df
```

### Functions to evaluate models

```python
def modelo_clas(data, objetivo, modelo, train_size= 0.8, cv = 5):
    """
    Function that allows training and evaluating a machine learning model 
    for classification problems.
    Args:
    data: the dataset used to train the model.
    objetivo: name of the target variable (str).
    modelo: a pre-specified model with its parameters.
    train_size: size of the dataset to be used for training, specified as a decimal fraction. Default value: 0.8.
    cv: the number of data splits to assess the model using cross-validation. Default value: 5.
    Return: direct printing of model evaluation metrics:
    Accuracy obtained through cross-validation.
    Confusion matrix.
    Classification report.
    """
    df = data.copy()
    X = df.drop([objetivo], axis = 1)
    y = df[[objetivo]]

    X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size = train_size,
                                                    random_state = 42)

    modelo.fit(X_train, y_train)
    pred = modelo.predict(X_test)

    cm = confusion_matrix(y_test, pred)
    cr = classification_report(y_test, pred)
    accuracy = accuracy_score(
            y_true    = y_test,
            y_pred    = pred,
            normalize = True
            )

    cv_scores = cross_val_score(modelo, X_train, y_train, cv=cv, scoring="accuracy") # Realizar validación cruzada y mostrar las métricas promedio
    print(f"Validación cruzada (exactitud): {100*cv_scores.mean()}% \n")
    print(f'Matriz de confusión: \n {cm} \n')
    print(f'Reporte de clasificación: \n {cr}')
```

### Functions for balancing categories

```python
def balancear_muestras(data, variable_objetivo):
    """
    Function that samples a dataset in such a way that each category of 
    a specified variable contains the same number of records.
    Args:
    data: dataset to be modified.
    target_variable: the variable that will determine the number of records.
    Return:
    balanced_df: a new dataset with records proportional to each category
    """
    conteo_muestras = data[variable_objetivo].value_counts()
    longitud_minima = conteo_muestras.min()
    df_balanceado = pd.DataFrame()
    for categoria in data[variable_objetivo].unique(): 
        filtro_categoria = data[data[variable_objetivo] == categoria] 
        muestras_categoria = filtro_categoria.sample(n=longitud_minima, random_state=42) 
        df_balanceado = pd.concat([df_balanceado, muestras_categoria], ignore_index=True)
    df_balanceado = df_balanceado.sample(frac=1, random_state=42)
    return df_balanceado
```

### Function to calculate the median by groups

```python
def calcular_mediana_por_grupo(group):
    """
    Function to calculate the median by gender and age for values other than 999 (Not Specified).
    Args:
    group: gender and age group.
    Return:
    Returns the median value of the specified group.
    """
    return group[group != 999].median()
```

### Functions to calculate Mutual Information

```python
def make_scores_clas(X, y):
    X = X.copy()
    mi_scores = mutual_info_classif(X, y, random_state=42)  
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)  
    mi_scores = mi_scores.sort_values(ascending=False)  
    return mi_scores
```

```python
def plot_scores(scores):
    scores = scores.sort_values(ascending=True)  
    width = range(len(scores)) 
    ticks = list(scores.index)  
    plt.barh(width, scores)  
    plt.yticks(width, ticks) 
    plt.title("Mutual Information Scores")
```

## Imputation of weight 'PESO' and height 'TALLA' values

```python
egresos_pt = egresos2.copy()
# Calculate an approximate weight based on the median for all weight values equal to 999
egresos_pt['PESO'] = egresos_pt.groupby(['SEXO_M', 'EDAD'])['PESO'].apply(lambda x: x.mask(x == 999, calcular_mediana_por_grupo(x)))
# Calculate an approximate height based on the median for all height values equal to 999
egresos_pt['TALLA'] = egresos_pt.groupby(['SEXO_M', 'EDAD'])['TALLA'].apply(lambda x: x.mask(x == 999, calcular_mediana_por_grupo(x)))
egresos_pt
```

```python
# Calculate the Body Mass Index 'IMC'
egresos2['IMC'] = egresos2['PESO']/(egresos2['TALLA']/100)**2
egresos_pt['IMC'] = egresos_pt['PESO']/(egresos_pt['TALLA']/100)**2
```
