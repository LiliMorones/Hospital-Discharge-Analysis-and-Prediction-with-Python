## Load datasets

```python
# Discharges
file = '../../Data/Databases/Before viz/egresos.csv'
egresos = pd.read_csv(file, on_bad_lines='skip')
egresos.head()
```

```python
egresos['EGRESO'] = pd.to_datetime(egresos['EGRESO'])
egresos['INGRE'] = pd.to_datetime(egresos['INGRE'])
```

```python
# Assign order to categorical data for age 'EDAD' and stay category 'DIAS_ESTA'
#Age
values = ['0 a 4', '5 a 14', '15 a 29', '30 a 44', '45 a 59', '60 o más']
egresos['CAT_EDAD'] = pd.Categorical(egresos['CAT_EDAD'], categories=values, ordered=True)
#Stay
values = ['1 o menos', '2 a 3', '4 a 7', '8 a 30', 'Más de 30']
egresos['CAT_DIASESTA'] = pd.Categorical(egresos['CAT_DIASESTA'], categories=values, ordered=True)
```

```python
#Affections
file = '../../Data/Databases/Before viz/afecciones.csv'
afecciones = pd.read_csv(file, on_bad_lines='skip')
afecciones.head()
```

```python
afecciones = afecciones[afecciones['ID'].isin(egresos['ID'])]
```

```python
#Procedures
file = '../../Data/Databases/Before viz/procedimientos.csv'
procedimientos = pd.read_csv(file, on_bad_lines='skip')
procedimientos.head()
```

```python
procedimientos = procedimientos[procedimientos['ID'].isin(egresos['ID'])]
```

```python
#Change category 'Uso emergente de u07' to 'COVID-19' 
afecciones['AFEC'] = afecciones['AFEC'].replace({'USO EMERGENTE DE U07': 'COVID-19'})
egresos['DIAG_INI'] = egresos['DIAG_INI'].replace({'USO EMERGENTE DE U07': 'COVID-19'})
egresos['AFECPRIN'] = egresos['AFECPRIN'].replace({'USO EMERGENTE DE U07': 'COVID-19'})
```

#### Medical procedures per year
Throughout all the years, the most performed procedure is diagnosis, and until 2020, surgeries were in second place. However, they were surpassed by therapeutic procedures, which doubled compared to previous years.

```python
# Select the columns of interest from the main table
anios = egresos[['ID', 'EGRESO']]

#Merge with procedures table
df = procedimientos.merge(anios, how='left', on='ID')
df['EGRESO'] = df['EGRESO'].dt.year
df = df[['TIPO', 'EGRESO']]

# remove mixed procedures as there are very few of them, and they are not present every year
df = df[~(df['TIPO']=='MIXTO')]

# Calculate the number of procedures per year and by type
count_data = df.groupby(['EGRESO', 'TIPO']).size().unstack()

# Line chart
plt.figure(figsize=(8, 6))
count_data.plot(kind='line', marker='o')
plt.title('Evolución de procedimientos médicos 2010-2022')
plt.xlabel('Año')
plt.ylabel('Número de Procedimientos')
plt.legend(title='Tipo', loc='center left', bbox_to_anchor=(1, 0.5))

# Show figure
plt.show()
```

The increase in therapeutic procedures coincides with the pandemic dates. We are going to check if, in fact, most of the patients with therapeutic procedures are COVID patients.<br>
We can see that the majority are natural births or patients with leukemia, but we do see that COVID comes into play, and it is interesting to note that there is an increase in cesarean sections (which probably also contributes to the increase seen in 2022 for surgical procedures).<br>
We see that other conditions such as metabolic disorders, fractures, and kidney disease have more significant numbers in the last year.<br>
In general, we can conclude that while COVID does have an influence, there is also an increase in treatments for many other conditions (keeping in mind that we are visualizing the number of patients receiving this care, not the number of therapeutic procedures for each of those patients).

```python
# Select only patients with therapeutic procedures (all years)
terapeuticos = procedimientos[procedimientos['TIPO']=='TERAPEUTICO']
df = egresos[egresos['ID'].isin(terapeuticos['ID'])]

# Filter the top 10 conditions from the year 2019 (before the increase in therapeutic procedures)
df1 = df[df['EGRESO'].dt.year==2019]
top10_df1 = df1['AFECPRIN'].value_counts().sort_values(ascending=False)[:10]

# Filter the top 10 conditions in 2022 (when therapeutic procedures doubled)
df2 = df[df['EGRESO'].dt.year==2022]
top10_df2 = df2['AFECPRIN'].value_counts().sort_values(ascending=False)[:10]

# Set figure
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Graph 1: Top 10 2019 Conditions
sns.barplot(
    x=top10_df1.values,
    y=[textwrap.fill(x, width=20) for x in top10_df1.index],
    color='darkcyan',
    ax=axes[0]
)

axes[0].set_title("Principales afecciones relacionadas a procedimientos terapéuticos (2019)")

# Show values in the right
for i, v in enumerate(top10_df1.values):
    axes[0].text(v + 1, i, str(v), color='black', va='center')

# Graph 2: Top 10 2022 Conditions
sns.barplot(
    x=top10_df2.values,
    y=[textwrap.fill(x, width=20) for x in top10_df2.index],
    color='cadetblue',
    ax=axes[1]
)

axes[1].set_title("Principales afecciones relacionadas a procedimientos terapéuticos (2022)")

# Show values in the right
for i, v in enumerate(top10_df2.values):
    axes[1].text(v + 1, i, str(v), color='black', va='center')

plt.tight_layout()
plt.show()
```

#### Evolution of the main conditions

The evolution from 2010 to 2022 of the 10 most common conditions in 2022 for hospital discharge is shown:<br>

1. Chronic kidney disease: Each year, more and more patients with this condition have been treated. Remember that Aguascalientes has one of the highest incidence rates worldwide for this disease. Cases have doubled in the last 10 years.<br>

2. Lymphoid leukemia: There is no clear trend, but it can be seen that the number of patients causing hospital discharges for this disease fluctuates between 1,250 and 1,550 each year.<br>

3. Single spontaneous birth (natural): There is a clear downward trend. On one hand, there are fewer women who want to give birth in public hospitals, but there is also the opposite... <br>

4. ...Cesarean section: It seems to be increasing, unlike natural birth. In this regard, the 2022 situational analysis of Aguascalientes mentions the increase in cases requiring cesarean section as a problem. It would be interesting to conduct a more in-depth analysis of this situation. <br>

5. Leg fracture: There has been a considerable increase in the number of cases in the last 2 years. It would be important to determine the main external cause of this condition to identify the root of the problem.<br>

6. Cholelithiasis: It appears that the number of cases decreased due to the pandemic and has increased again in the last year. It is likely that patients preferred admission to a hospital where COVID cases were not treated. <br>

7. COVID-19: Despite the pandemic starting in 2020, we see the highest number of cases in 2021, even more cases than in 2020 in 2022. Pending confirmation, but it seems that among the hospitals we are analyzing, only Hospital Hidalgo was a COVID hospital, and its capacity to provide care likely increased precisely in 2021. <br>

8. Forearm fracture: Like with the leg fracture, there has been an increase in recent years. However, the number of cases each year has been fluctuating between 200 and 300. We should check whether the number of cases increases again next year or decreases. <br>

9. Glucosaminoglycan metabolism disorder: Although there are not many cases yet, there is a clear upward trend. It is necessary to investigate why this situation is occurring and whether it is just an increase in attention to this condition or an increase in the number of cases. <br>

10. Femur fracture: It follows the same pattern as the other two fracture cases."

```python
df = egresos.copy()
df['ANIO'] = df['EGRESO'].dt.year
df_2022 = df[df['ANIO'] == 2022]
top_10_afecprin = df_2022['AFECPRIN'].value_counts().nlargest(10).index.tolist()

#Create the dictionary where the key is the primary condition, and the values are lists with the number of cases for each year
data_dict = {}
for afecprin in top_10_afecprin:
    data_dict[afecprin] = [df[(df['ANIO'] == year) & (df['AFECPRIN'] == afecprin)].shape[0] for year in range(2010, 2023)]

#Create the figure with 5 rows and 2 columns
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(12, 18))

# Iterate over the primary conditions and their corresponding graphs.
for i, afecprin in enumerate(top_10_afecprin):
    # years and the corresponding data for the primary condition from the dictionary
    years = list(range(2010, 2023))
    data = data_dict[afecprin]

    # position of the graph in the figure
    row = i // 2
    col = i % 2

    # evolution of the primary condition
    axes[row, col].plot(years, data, marker='o')
    axes[row, col].set_title(afecprin)
    axes[row, col].set_xlabel('')
    axes[row, col].set_ylabel('Número de egresos')

# Adjust subplots
plt.tight_layout()

#Show figure
plt.show()
```
