## Import libraries

```python
import pandas as pd
import numpy as np
```

## Load databases from Procedures

```python
file = '../../Data/Databases/Data Integration/procedimientos2010_2013.csv'
p10_13 = pd.read_csv(file, on_bad_lines='skip')
p10_13.head()
```

```python
file = '../../Data/Databases/Data Integration/procedimientos2014_2015.csv'
p14_15 = pd.read_csv(file, on_bad_lines='skip')
p14_15.head()
```

```python
file = '../../Data/Databases/Data Integration/procedimientos2016_2017.csv'
p16_17 = pd.read_csv(file, on_bad_lines='skip')
p16_17.head()
```

```python
file = '../../Data/Databases/Data Integration/procedimientos2018_2019.csv'
p18_19 = pd.read_csv(file, on_bad_lines='skip')
p18_19.head()
```

```python
file = '../../Data/Databases/Data Integration/procedimientos2020_2021.csv'
p20_21 = pd.read_csv(file, on_bad_lines='skip')
p20_21.head()
```

```python
file = '../../Data/Databases/Data Integration/procedimientos2022.csv'
p22 = pd.read_csv(file, on_bad_lines='skip')
p22.head()
```

## Integrate databases

```python
procedimientos = pd.concat([p10_13, p14_15, p16_17, p18_19, p20_21, p22])
procedimientos.head()
```

## Rename ID column

```python
procedimientos.rename(columns={'ID2':'ID'},inplace=True)
```

```python
procedimientos = procedimientos[['ID','NUMPROMED','TIPO','PROMED','ANEST','QUIROF','QH','QM']]
procedimientos.head()
```

## Export data to csv

```python
from pathlib import Path
filepath = Path('../../Data/Databases/Data Integration/procedimientos2010_2022.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
procedimientos.to_csv(filepath, index = False)
```
