```python
import pandas as pd
import numpy as np
```

## Carga de bases de datos Egresos

```python
file = '../../Registros/Bases de datos/Egresos/EGRESOS_2018.csv'
egresos2018 = pd.read_csv(file, on_bad_lines='skip')
egresos2018.head()
```

```python
file = '../../Registros/Bases de datos/Egresos/EGRESOS_2019.txt'
egresos2019 = pd.read_table(file, delimiter = ',')
egresos2019.head()
```

## Filtrado de registros Aguascalientes

```python
#Conservar registros donde la entidad es Aguascalientes (1)
egresos2018 = egresos2018[egresos2018['ENTIDAD'] == 1]
egresos2019 = egresos2019[egresos2019['ENTIDAD'] == 1]
#Conservar registros donde el hospital sea de Ags (excluyendo Hospital de la Mujer y Hospital Psiquiátrico)
ags = ['ASSSA000404', 'ASSSA000030', 'ASSSA000614', 'ASSSA000655', 'ASSSA000025', 'ASSSA001331']
egresos2019 = egresos2019[egresos2019['CLUES'].isin(ags)]
egresos2018 = egresos2018[egresos2018['CLUES'].isin(ags)]
```

## Carga de bases de datos Afecciones

```python
file = '../Registros/2019/AFECCIONES_2019.txt'
afecciones2019 = pd.read_table(file, delimiter = ',')
afecciones2019.head()
```

```python
file = '../Registros/2018/AFECCIONES_2018.csv'
afecciones2018 = pd.read_csv(file, on_bad_lines='skip')
afecciones2018.head()
```
