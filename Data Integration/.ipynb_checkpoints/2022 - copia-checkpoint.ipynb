## Importar librerías

```python
import pandas as pd
import numpy as np
```

## Carga de bases de datos Egresos

```python
file = '../../Registros/Bases de datos/Egresos/EGRESOS_2022.txt'
egresos2022 = pd.read_table(file, delimiter = '|')
egresos2022.head()
```

## Filtrado de registros Aguascalientes

```python
#Conservar registros donde la entidad es Aguascalientes (1)
egresos2022 = egresos2022[egresos2022['ENTIDAD'] == 1]
#Conservar registros donde el hospital sea de Ags (excluyendo Hospital de la Mujer y Hospital Psiquiátrico)
ags = ['ASSSA000404', 'ASSSA000030', 'ASSSA000614', 'ASSSA000655', 'ASSSA000025', 'ASSSA001331']
egresos2022 = egresos2022[egresos2022['CLUES'].isin(ags)]
```

## Cambiar tipo de dato

```python
# Cambiar tipo de dato de 'ID' a entero
egresos2022['ID'] = egresos2022['ID'].astype('int64')
egresos2022.info()
```

## Reasignación de ID's Egresos

```python
#2021: termina en 298042
#Crear una columna para numerar los registros en orden por año
egresos2022['ID2'] = np.arange(298043,298043 + len(egresos2022))
```

```python
#Crear datasets con el ID original y el nuevo ID
IDs_2022 = egresos2022[['ID','ID2']]
```

## Cambiar clasificación CVEEDAD y PROCED

```python
egresos2022['CVEEDAD'].value_counts()
```

```python
egresos2022['CVEEDAD'] = egresos2022['CVEEDAD'].replace({2:0, 3:1, 4:2, 5:3})
```

```python
egresos2022['PROCED'].value_counts()
```

```python
egresos2022['PROCED'] = egresos2022['PROCED'].replace({4:6,5:4})
```

## Carga de bases de datos Afecciones

```python
file = '../../Registros/Bases de datos/Afecciones/AFECCIONES_2022.txt'
afecciones2022 = pd.read_table(file, delimiter = '|')
afecciones2022.head()
```

## Relación afecciones con nuevo ID

```python
#Combinar la tabla afecciones con la tabla IDs para tener el valor del nuevo ID correspondiente
afec2022 = afecciones2022.merge(IDs_2022, how = 'inner', on = 'ID')
```

```python
#Comprobar que se anexa ID2
afec2022.head()
```

```python
#Comprobar que se asigna correctamente el ID
egresos2022[egresos2022['ID']==1028846]
```

## Cargar bases de datos Procedimientos

```python
file = '../../Registros/Bases de datos/Procedimientos/PROCEDIMIENTOS_2022.txt'
procedimientos2022 = pd.read_table(file, delimiter = '|')
procedimientos2022.head()
```

```python
#Separar la columna de tiempo quirófano en horas y minutos para que coincida con los demás años
quirofano = procedimientos2022["tiempo_quirofano"].str.split(':', expand=True)
quirofano.columns = ['QH', 'QM']
procedimientos2022 = pd.concat([procedimientos2022, quirofano], axis=1)
procedimientos2022.drop(['tiempo_quirofano'], axis = 1, inplace = True)
procedimientos2022.head()
```

## Relación procedimientos con nuevo ID

```python
#Combinar la tabla procedimientos con la tabla ID
proc2022 = procedimientos2022.merge(IDs_2022, how = 'inner', on = 'ID')
```

```python
#Comprobar que se anexa ID2
proc2022.head()
```

```python
#Comprobar que se asigna correctamente el ID
egresos2022[egresos2022['ID']==1028842]
```

## Filtrar columnas

```python
#Conservar solo las variables que son de nuestro interés
v_keep = ['ID2', 'CLUES', 'EGRESO', 'INGRE', 'DIAS_ESTA', 'CVEEDAD','EDAD', 'NACIOEN', 'SEXO', 'PESO', 'TALLA', 'DERHAB',
         'MUNIC', 'LOC', 'TIPSERV', 'PROCED', 'MOTEGRE', 'DIAG_INI', 'AFECPRIN', 'VEZ', 'INFEC', 'CAUSAEXT']
egresos2022 = egresos2022[v_keep]
```

```python
#Eliminar la columna del ID original de la tabla afecciones
afec2022.drop(['ID'], axis = 1, inplace = True)
```

```python
#Eliminar la columna del ID original de la tabla procedimientos
proc2022.drop(['ID'], axis = 1, inplace = True)
```
