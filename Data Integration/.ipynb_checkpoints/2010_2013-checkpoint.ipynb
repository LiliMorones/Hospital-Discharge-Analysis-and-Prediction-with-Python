## Importar Librerías

```python
import pandas as pd
import numpy as np
```

## Carga de bases de datos Egresos

```python
file = '../Registros/Bases de datos/Egresos/EGRESOS_2010.csv'
egresos2010 = pd.read_csv(file, on_bad_lines='skip')
egresos2010.head()
```

```python
file = '../Registros/Bases de datos/Egresos/EGRESOS_2011.csv'
egresos2011 = pd.read_csv(file, on_bad_lines='skip')
egresos2011.head()
```

```python
file = '../Registros/Bases de datos/Egresos/EGRESOS_2012.csv'
egresos2012 = pd.read_csv(file, on_bad_lines='skip')
egresos2012.head()
```

```python
file = '../Registros/2013/EGRESOS_2013.csv'
egresos2013 = pd.read_csv(file, on_bad_lines='skip')
egresos2013.head()
```

## Filtrado de registros Aguascalientes

```python
egresos2010 = egresos2010[egresos2010['ENTIDAD'] == 1]
egresos2011 = egresos2011[egresos2011['ENTIDAD'] == 1]
egresos2012 = egresos2012[egresos2012['ENTIDAD'] == 1]
egresos2013 = egresos2013[egresos2013['ENTIDAD'] == 1]
ags = ['ASSSA000404', 'ASSSA000030', 'ASSSA000614', 'ASSSA000655', 'ASSSA000025', 'ASSSA001331']
egresos2010 = egresos2010[egresos2010['CLUES'].isin(ags)]
egresos2011 = egresos2011[egresos2011['CLUES'].isin(ags)]
egresos2012 = egresos2012[egresos2012['CLUES'].isin(ags)]
egresos2013 = egresos2013[egresos2013['CLUES'].isin(ags)]
print(len(egresos2010))
print(len(egresos2011))
print(len(egresos2012))
print(len(egresos2013))
```

```python
egresos2013.head()
```

## Reasignación de ID's Egresos

```python
egresos2010['ID2'] = np.arange(1, 1 + len(egresos2010))
egresos2011['ID2'] = np.arange(26007,26007 + len(egresos2011))
egresos2012['ID2'] = np.arange(53694,53694 + len(egresos2012))
egresos2013['ID2'] = np.arange(81578,81578 + len(egresos2013))
```

```python
print(egresos2010['ID2'].max())
print(egresos2011['ID2'].min())
print(egresos2011['ID2'].max())
print(egresos2012['ID2'].min())
print(egresos2012['ID2'].max())
print(egresos2013['ID2'].min())
print(egresos2013['ID2'].max())
```

```python
IDs_2010 = egresos2010[['ID','ID2']]
IDs_2011 = egresos2011[['ID','ID2']]
IDs_2012 = egresos2012[['ID','ID2']]
IDs_2013 = egresos2013[['ID','ID2']]
```
