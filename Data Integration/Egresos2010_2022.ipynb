## Data Integration

```python
egresos = pd.concat([e10_13, e14_15, e16_17, e18_19, e20_21, e22])
egresos.head()
```

## Rename ID column

```python
egresos.rename(columns={'ID2':'ID'},inplace=True)
egresos.head()
```

## Change data types

```python
egresos.info()
```

```python
egresos['DIAS_ESTA'] = egresos['DIAS_ESTA'].astype(int)
```

```python
egresos['EDAD'] = egresos['EDAD'].astype(int)
```

```python
egresos['SEXO'] = egresos['SEXO'].astype(int)
```

```python
egresos['MUNIC'] = egresos['MUNIC'].astype(int)
```

```python
egresos['LOC'] = egresos['LOC'].astype(int)
```

## Export data to csv

```python
from pathlib import Path 
filepath = Path('../../Data/Databases/Data Integration/egresos2010_2022.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
egresos.to_csv(filepath, index = False)
```
