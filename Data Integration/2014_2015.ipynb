# 2014 - 2015 Data Integration

## Import libraries

```python
import pandas as pd
import numpy as np
```

## Load Databases from Discharges

```python
file = '../../Data/Databases/Discharges/EGRESOS_2014.csv'
egresos2014 = pd.read_csv(file, on_bad_lines='skip')
egresos2014.head()
```

```python
file = '../../Data/Databases/Discharges/EGRESOS_2015.csv'
egresos2015 = pd.read_csv(file, on_bad_lines='skip')
egresos2015.head()
```

## Keep only records from Aguascalientes

```python
#Keep records where 'ENTIDAD' = Aguascalientes (1)
egresos2014 = egresos2014[egresos2014['ENTIDAD'] == 1]
egresos2015 = egresos2015[egresos2015['ENTIDAD'] == 1]
#Keep records for hospitals in Ags (excluding 'Hospital de la Mujer' and 'Hospital PsiquiÃ¡trico')
ags = ['ASSSA000404', 'ASSSA000030', 'ASSSA000614', 'ASSSA000655', 'ASSSA000025', 'ASSSA001331']
egresos2014 = egresos2014[egresos2014['CLUES'].isin(ags)]
egresos2015 = egresos2015[egresos2015['CLUES'].isin(ags)]
```

## Reassign discharge id's

```python
#2013 ended with 108362
print(108362+len(egresos2014))
```

```python
#Number the records in chronological order by year
egresos2014['ID2'] = np.arange(108363,108363 + len(egresos2014))
egresos2015['ID2'] = np.arange(135455,135455 + len(egresos2015))
```

```python
#Create datasets with the original ID and the new ID
IDs_2014 = egresos2014[['ID','ID2']]
IDs_2015 = egresos2015[['ID','ID2']]
```

## Replace MOTEGRE key

```python
##Change a key, replace 6 with 7 in all records to match the classification of the other years (2018-2022)
egresos2014['MOTEGRE'] = egresos2014['MOTEGRE'].replace({6:7})
egresos2015['MOTEGRE'] = egresos2015['MOTEGRE'].replace({6:7})
```

## Load Databases from Affections

```python
file = '../../Data/Databases/Affections/AFECCIONES_2014.csv'
afecciones2014 = pd.read_csv(file, on_bad_lines='skip')
afecciones2014.head()
```

```python
file = '../../Data/Databases/Affections/AFECCIONES_2015.csv'
afecciones2015 = pd.read_csv(file, on_bad_lines='skip')
afecciones2015.head()
```

```python
#Drop NULL values from ID - 2015 and change the data type to int
afecciones2015.dropna(inplace = True)
afecciones2015['ID'] = afecciones2015['ID'].astype('int64')
```

## Associate affections with the new ID column

```python
afec2014 = afecciones2014.merge(IDs_2014, how = 'inner', on = 'ID')
afec2015 = afecciones2015.merge(IDs_2015, how = 'inner', on = 'ID')
afec2015.head()
```

## Load Databases from Procedures

```python
file = '../../Data/Databases/Procedures/PROCEDIMIENTOS_2014.csv'
procedimientos2014 = pd.read_csv(file, on_bad_lines='skip')
procedimientos2014.head()
```

```python
file = '../../Data/Databases/Procedures/PROCEDIMIENTOS_2015.csv'
procedimientos2015 = pd.read_csv(file, on_bad_lines='skip')
procedimientos2015.head()
```

```python
#Drop NULL values from ID - 2015 and change the data type to int
procedimientos2015.dropna(subset = ['ID'],inplace = True)
procedimientos2015['ID'] = procedimientos2015['ID'].astype('int64')
```

## Associate procedures with the new ID column

```python
proc2014 = procedimientos2014.merge(IDs_2014, how = 'inner', on = 'ID')
proc2015 = procedimientos2015.merge(IDs_2015, how = 'inner', on = 'ID')
proc2015.head()
```

## Keep variables of interest

```python
v_keep = ['ID2', 'CLUES', 'EGRESO', 'INGRE', 'DIAS_ESTA', 'CVEEDAD','EDAD', 'NACIOEN', 'SEXO', 'PESO', 'TALLA', 'DERHAB',
         'MUNIC', 'LOC', 'TIPSERV', 'PROCED', 'MOTEGRE', 'DIAG_INI', 'AFECPRIN', 'VEZ', 'INFEC', 'CAUSAEXT']
egresos2014 = egresos2014[v_keep]
egresos2015 = egresos2015[v_keep]
```

```python
#Drop original ID column from Affections
afec2014.drop(['ID'], axis = 1, inplace = True)
afec2015.drop(['ID'], axis = 1, inplace = True)
```

```python
#Drop original ID column from Procedures
proc2014.drop(['ID'], axis = 1, inplace = True)
proc2015.drop(['ID'], axis = 1, inplace = True)
```

## Integrate Databases

```python
#Discharges
egresos2014_2015 = pd.concat([egresos2014, egresos2015])
egresos2014_2015
```

```python
#Affections
afecciones2014_2015 = pd.concat([afec2014, afec2015])
afecciones2014_2015
```

```python
#Procedures
procedimientos2014_2015 = pd.concat([proc2014, proc2015])
procedimientos2014_2015
```

## Exportar Data to CSV

```python
from pathlib import Path  
#Discharges
filepath = Path('../../Data/Databases/Data Integration/egresos2014_2015.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
egresos2014_2015.to_csv(filepath, index = False)
```

```python
#Affections
filepath = Path('../../Data/Databases/Data Integration/afecciones2014_2015.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
afecciones2014_2015.to_csv(filepath, index = False)
```

```python
#Procedures
filepath = Path('../../Data/Databases/Data Integration/procedimientos2014_2015.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
procedimientos2014_2015.to_csv(filepath, index = False)
```
