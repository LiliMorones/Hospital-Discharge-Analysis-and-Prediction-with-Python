# 2020 - 2021 Data Integration

## Import libraries

```python
import pandas as pd
import numpy as np
```

## Load Databases from Discharges

```python
file = '../../Data/Databases/Discharges/EGRESOS_2020.txt'
egresos2020 = pd.read_table(file, delimiter = '|')
egresos2020.head()
```

```python
file = '../../Data/Databases/Discharges/EGRESOS_2021.txt'
egresos2021 = pd.read_table(file, delimiter = '|')
egresos2021.head()
```

## Keep only records from Aguascalientes

```python
#Keep records where 'ENTIDAD' = Aguascalientes (1)
egresos2020 = egresos2020[egresos2020['ENTIDAD'] == 1]
egresos2021 = egresos2021[egresos2021['ENTIDAD'] == 1]
#Keep records for hospitals in Ags (excluding 'Hospital de la Mujer' and 'Hospital PsiquiÃ¡trico')
ags = ['ASSSA000404', 'ASSSA000030', 'ASSSA000614', 'ASSSA000655', 'ASSSA000025', 'ASSSA001331']
egresos2020 = egresos2020[egresos2020['CLUES'].isin(ags)]
egresos2021 = egresos2021[egresos2021['CLUES'].isin(ags)]
```

```python
#Change ID data type to int - 2021
egresos2021['ID'] = egresos2021['ID'].astype('int64')
```

## Reassign discharge id's

```python
#2019 ended with 257455
print(257455+len(egresos2020))
```

```python
#Number the records in chronological order by year
egresos2020['ID2'] = np.arange(257456,257456 + len(egresos2020))
egresos2021['ID2'] = np.arange(276898,276898 + len(egresos2021))
```

```python
#Create datasets with the original ID and the new ID
IDs_2020 = egresos2020[['ID','ID2']]
IDs_2021 = egresos2021[['ID','ID2']]
```

## Replace CVEEDAD and PROCED keys

```python
# Update keys to match previous years (2010-2019)
egresos2020['CVEEDAD'] = egresos2020['CVEEDAD'].replace({2:0, 3:1, 4:2, 5:3})
egresos2021['CVEEDAD'] = egresos2021['CVEEDAD'].replace({2:0, 3:1, 4:2, 5:3})
egresos2020['PROCED'] = egresos2020['PROCED'].replace({5:4})
egresos2021['PROCED'] = egresos2021['PROCED'].replace({4:6, 5:4})
```

## Load Databases from Affections

```python
file = '../../Data/Databases/Affections/AFECCIONES_2020.txt'
afecciones2020 = pd.read_table(file, delimiter = '|')
afecciones2020.head()
```

```python
#Rename columns
afecciones2020.rename(columns={'cod_cie_comorbilidad':'AFEC', 'numero_comorbilidad':'NUMAFEC'},inplace=True)
afecciones2020.head()
```

```python
file = '../../Data/Databases/Affections/AFECCIONES_2021.txt'
afecciones2021 = pd.read_table(file, delimiter = '|')
afecciones2021.head()
```

## Associate affections with the new ID column

```python
afec2020 = afecciones2020.merge(IDs_2020, how = 'inner', on = 'ID')
afec2021 = afecciones2021.merge(IDs_2021, how = 'inner', on = 'ID')
afec2021.head()
```

## Load Databases from Procedures

```python
file = '../../Data/Databases/Procedures/PROCEDIMIENTOS_2020.txt'
procedimientos2020 = pd.read_table(file, delimiter = '|')
procedimientos2020.head()
```

```python
#Split 'tiempo_quirofano' into 2 columns: 'QH' for hours and 'QM' for minutes
quirofano = procedimientos2020['tiempo_quirofano'].str.split(':', expand=True)
quirofano.columns = ['QH', 'QM']
procedimientos2020 = pd.concat([procedimientos2020, quirofano], axis=1)
procedimientos2020.drop(['tiempo_quirofano'], axis = 1, inplace = True)
procedimientos2020.head()
```

```python
file = '../../Data/Databases/Procedures/PROCEDIMIENTOS_2021.txt'
procedimientos2021 = pd.read_table(file, delimiter = '|')
procedimientos2021.head()
```

```python
#Split 'tiempo_quirofano' into 2 columns: 'QH' for hours and 'QM' for minutes
quirofano = procedimientos2021["tiempo_quirofano"].str.split(':', expand=True)
quirofano.columns = ['QH', 'QM']
procedimientos2021 = pd.concat([procedimientos2021, quirofano], axis=1)
procedimientos2021.drop(['tiempo_quirofano'], axis = 1, inplace = True)
procedimientos2021.head()
```

## Associate procedures with the new ID column

```python
proc2020 = procedimientos2020.merge(IDs_2020, how = 'inner', on = 'ID')
proc2021 = procedimientos2021.merge(IDs_2021, how = 'inner', on = 'ID')
proc2021.head()
```

## Keep variables of interest

```python
v_keep = ['ID2', 'CLUES', 'EGRESO', 'INGRE', 'DIAS_ESTA', 'CVEEDAD','EDAD', 'NACIOEN', 'SEXO', 'PESO', 'TALLA', 'DERHAB',
         'MUNIC', 'LOC', 'TIPSERV', 'PROCED', 'MOTEGRE', 'DIAG_INI', 'AFECPRIN', 'VEZ', 'INFEC', 'CAUSAEXT']
egresos2020 = egresos2020[v_keep]
egresos2021 = egresos2021[v_keep]
```

```python
#Drop original ID column from Affections
afec2020.drop(['ID'], axis = 1, inplace = True)
afec2021.drop(['ID'], axis = 1, inplace = True)
```

```python
#Drop original ID column from Procedures
proc2020.drop(['ID'], axis = 1, inplace = True)
proc2021.drop(['ID'], axis = 1, inplace = True)
```

## Integrate Databases

```python
#Discharges
egresos2020_2021 = pd.concat([egresos2020, egresos2021])
egresos2020_2021
```

```python
#Affections
afecciones2020_2021 = pd.concat([afec2020, afec2021])
afecciones2020_2021
```

```python
#Procedures
procedimientos2020_2021 = pd.concat([proc2020, proc2021])
procedimientos2020_2021
```
