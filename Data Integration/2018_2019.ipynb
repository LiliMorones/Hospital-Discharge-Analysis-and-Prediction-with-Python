# 2018 - 2019 Data Integration

## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Load Databases from Discharges

```python
file = '../../Data/Databases/Discharges/EGRESOS_2018.csv'
egresos2018 = pd.read_csv(file, on_bad_lines='skip')
egresos2018.head()
```

```python
file = '../../Data/Databases/Discharges/EGRESOS_2019.txt'
egresos2019 = pd.read_table(file, delimiter = ',')
egresos2019.head()
```

## Keep only records from Aguascalientes

```python
#Keep records where 'ENTIDAD' = Aguascalientes (1)
egresos2018 = egresos2018[egresos2018['ENTIDAD'] == 1]
egresos2019 = egresos2019[egresos2019['ENTIDAD'] == 1]
#Keep records for hospitals in Ags (excluding 'Hospital de la Mujer' and 'Hospital Psiquiátrico')
ags = ['ASSSA000404', 'ASSSA000030', 'ASSSA000614', 'ASSSA000655', 'ASSSA000025', 'ASSSA001331']
egresos2019 = egresos2019[egresos2019['CLUES'].isin(ags)]
egresos2018 = egresos2018[egresos2018['CLUES'].isin(ags)]
```

## Reassign discharge id's

```python
#2017 ended with 213268
print(213268+len(egresos2018))
```

```python
#Number the records in chronological order by year
egresos2018['ID2'] = np.arange(213269,213269 + len(egresos2018))
egresos2019['ID2'] = np.arange(235170, 235170 + len(egresos2019))
```

```python
#Create datasets with the original ID and the new ID
IDs_2018 = egresos2018[['ID','ID2']]
IDs_2019 = egresos2019[['ID','ID2']]
```

## Load Databases from Affections

```python
file = '../../Data/Databases/Affections/AFECCIONES_2018.csv'
afecciones2018 = pd.read_csv(file, on_bad_lines='skip')
afecciones2018.head()
```

```python
file = '../../Data/Databases/Affections/AFECCIONES_2019.txt'
afecciones2019 = pd.read_table(file, delimiter = ',')
afecciones2019.head()
```

## Associate affections with the new ID column

```python
afec2018 = afecciones2018.merge(IDs_2018, how = 'inner', on = 'ID')
afec2019 = afecciones2019.merge(IDs_2019, how = 'inner', on = 'ID')
afec2018.head()
```

## Load Databases from Procedures

```python
file = '../../Data/Databases/Procedures/PROCEDIMIENTOS_2018.csv'
procedimientos2018 = pd.read_csv(file, on_bad_lines='skip')
procedimientos2018.head()
```

```python
file = '../../Data/Databases/Procedures/PROCEDIMIENTOS_2019.txt'
procedimientos2019 = pd.read_table(file, delimiter = ',')
procedimientos2019.head()
```

## Associate procedures with the new ID column

```python
proc2019 = procedimientos2019.merge(IDs_2019, how = 'inner', on = 'ID')
proc2018 = procedimientos2018.merge(IDs_2018, how = 'inner', on = 'ID')
proc2018.head()
```

## Keep variables of interest

```python
v_keep = ['ID2', 'CLUES', 'EGRESO', 'INGRE', 'DIAS_ESTA', 'CVEEDAD','EDAD', 'NACIOEN', 'SEXO', 'PESO', 'TALLA', 'DERHAB',
         'MUNIC', 'LOC', 'TIPSERV', 'PROCED', 'MOTEGRE', 'DIAG_INI', 'AFECPRIN', 'VEZ', 'INFEC', 'CAUSAEXT']
egresos2019 = egresos2019[v_keep]
egresos2018 = egresos2018[v_keep]
```

```python
#Drop original ID column from Affections
afec2019.drop(['ID'], axis = 1, inplace = True)
afec2018.drop(['ID'], axis = 1, inplace = True)
```

```python
#Drop original ID column from Procedures
proc2019.drop(['ID'], axis = 1, inplace = True)
proc2018.drop(['ID'], axis = 1, inplace = True)
```

## Integrate Databases

```python
#Discharges
egresos2018_2019 = pd.concat([egresos2018, egresos2019])
egresos2018_2019
```

```python
#Affections
afecciones2018_2019 = pd.concat([afec2018, afec2019])
afecciones2018_2019
```

```python
#Procedures
procedimientos2018_2019 = pd.concat([proc2018, proc2019])
procedimientos2018_2019
```

## Export data to CSV

```python
from pathlib import Path  
#Discharges
filepath = Path('../../Data/Databases/Data Integration/egresos2018_2019.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
egresos2018_2019.to_csv(filepath, index = False)
```

```python
#Affections
filepath = Path('../../Data/Databases/Data Integration/afecciones2018_2019.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
afecciones2018_2019.to_csv(filepath, index = False)
```

```python
#Procedures
filepath = Path('../../Data/Databases/Data Integration/procedimientos2018_2019.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
procedimientos2018_2019.to_csv(filepath, index = False)
```
