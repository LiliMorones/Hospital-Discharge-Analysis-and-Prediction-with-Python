# 2016 - 2017 Data Integration

## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Load Databases from Discharges

```python
file = '../../Data/Databases/Discharges/EGRESOS_2016.csv'
egresos2016 = pd.read_csv(file, on_bad_lines='skip')
egresos2016.head()
```

```python
file = '../../Data/Databases/Discharges/EGRESOS_2017.csv'
egresos2017 = pd.read_csv(file, on_bad_lines='skip', delimiter = '|')
egresos2017.head()
```

## Keep only records from Aguascalientes

```python
#Keep records where 'ENTIDAD' = Aguascalientes (1)
egresos2016 = egresos2016[egresos2016['ENTIDAD'] == 1]
egresos2017 = egresos2017[egresos2017['ENTIDAD'] == 1]
#Keep records for hospitals in Ags (excluding 'Hospital de la Mujer' and 'Hospital Psiquiátrico')
ags = ['ASSSA000404', 'ASSSA000030', 'ASSSA000614', 'ASSSA000655', 'ASSSA000025', 'ASSSA001331']
egresos2016 = egresos2016[egresos2016['CLUES'].isin(ags)]
egresos2017 = egresos2017[egresos2017['CLUES'].isin(ags)]
```

## Reassign discharge id's

```python
#2015 ends with 161766
print(161766+len(egresos2016))
```

```python
#Number the records in chronological order by year
egresos2016['ID2'] = np.arange(161767,161767 + len(egresos2016))
egresos2017['ID2'] = np.arange(187634,187634 + len(egresos2017))
```

```python
#Create datasets with the original ID and the new ID
IDs_2016 = egresos2016[['ID','ID2']]
IDs_2017 = egresos2017[['ID','ID2']]
```

## Replace MOTEGRE key

```python
##Change a key, replace 6 with 7 in all records to match the classification of the other years (2018-2022)
egresos2016['MOTEGRE'] = egresos2016['MOTEGRE'].replace({6:7})
egresos2017['MOTEGRE'] = egresos2017['MOTEGRE'].replace({7:9})
```

## Load Databases from Affections

```python
file = '../../Data/Databases/Affections/AFECCIONES_2016.csv'
afecciones2016 = pd.read_csv(file, on_bad_lines='skip')
afecciones2016.head()
```

```python
file = '../../Data/Databases/Affections/AFECCIONES_2017.csv'
afecciones2017 = pd.read_csv(file, on_bad_lines='skip', encoding = 'cp1252', delimiter = '|')
afecciones2017.head()
```

## Associate affections with the new ID column

```python
afec2016 = afecciones2016.merge(IDs_2016, how = 'inner', on = 'ID')
afec2017 = afecciones2017.merge(IDs_2017, how = 'inner', on = 'ID')
afec2017.head()
```

## Load Databases from Procedures

```python
file = '../../Data/Databases/Procedures/PROCEDIMIENTOS_2016.csv'
procedimientos2016 = pd.read_csv(file, on_bad_lines='skip')
procedimientos2016.head()
```

```python
file = '../../Data/Databases/Procedures/PROCEDIMIENTOS_2017.csv'
procedimientos2017 = pd.read_csv(file, on_bad_lines='skip', delimiter = '|')
procedimientos2017.head()
```

## Associate procedures with the new ID column

```python
proc2016 = procedimientos2016.merge(IDs_2016, how = 'inner', on = 'ID')
proc2017 = procedimientos2017.merge(IDs_2017, how = 'inner', on = 'ID')
proc2017.head()
```

## Keep variables of interest

```python
v_keep = ['ID2', 'CLUES', 'EGRESO', 'INGRE', 'DIAS_ESTA', 'CVEEDAD','EDAD', 'NACIOEN', 'SEXO', 'PESO', 'TALLA', 'DERHAB',
         'MUNIC', 'LOC', 'TIPSERV', 'PROCED', 'MOTEGRE', 'DIAG_INI', 'AFECPRIN', 'VEZ', 'INFEC', 'CAUSAEXT']
egresos2016 = egresos2016[v_keep]
egresos2017 = egresos2017[v_keep]
```

```python
#Drop original ID column from Affections
afec2017.drop(['ID'], axis = 1, inplace = True)
afec2016.drop(['ID'], axis = 1, inplace = True)
```

```python
#Drop original ID column from Procedures
proc2017.drop(['ID'], axis = 1, inplace = True)
proc2016.drop(['ID'], axis = 1, inplace = True)
```

## Integrate Databases

```python
#Discharges
egresos2016_2017 = pd.concat([egresos2016, egresos2017])
egresos2016_2017
```

```python
#Affections
afecciones2016_2017 = pd.concat([afec2016, afec2017])
afecciones2016_2017
```

```python
#Procedures
procedimientos2016_2017 = pd.concat([proc2016, proc2017])
procedimientos2016_2017
```

## Exportar data to CSV

```python
from pathlib import Path  
#Discharges
filepath = Path('../../Data/Databases/Data Integration/egresos2016_2017.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
egresos2016_2017.to_csv(filepath, index = False)
```

```python
#Affections
filepath = Path('../../Data/Databases/Data Integration/afecciones2016_2017.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
afecciones2016_2017.to_csv(filepath, index = False)
```

```python
#Procedures
filepath = Path('../../Data/Databases/Data Integration/procedimientos2016_2017.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
procedimientos2016_2017.to_csv(filepath, index = False)
```
