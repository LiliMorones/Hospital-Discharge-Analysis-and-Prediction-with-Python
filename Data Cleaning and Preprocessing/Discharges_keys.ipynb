# Discharges - Changing keys to their corresponding catalog descriptions 

## Import libraries

```python
import pandas as pd
import numpy as np
```

## Load dataset

```python
#Discharges datasets
file = '../../Data/Databases/Data Integration/egresos2010_2022.csv'
egresos = pd.read_csv(file, on_bad_lines='skip')
egresos.head()
```

```python
#Affections keys
file = '../../Data/Keys/CLAVES_AFEC.xlsx'
claves = pd.read_excel(file)
claves.head()
```

## Change date format

```python
egresos['EGRESO'] = pd.to_datetime(egresos['EGRESO'], dayfirst=True)
egresos['INGRE'] = pd.to_datetime(egresos['INGRE'], dayfirst=True)
```

## Replace keys

```python
#Hospital keys
egresos['CLUES']= egresos['CLUES'].replace(['ASSSA000025', 'ASSSA000030', 'ASSSA000404', 'ASSSA000614', 
                                            'ASSSA000655', 'ASSSA001331'],
                                           ['CENTENARIO HOSPITAL MIGUEL HIDALGO', 'HOSPITAL GENERAL TERCER MILENIO', 
                                            'HOSPITAL GENERAL CALVILLO', 'HOSPITAL GENERAL PABELLÓN DE ARTEAGA', 
                                            'HOSPITAL GENERAL RINCÓN DE ROMOS', 'CENTENARIO HOSPITAL MIGUEL HIDALGO'])
```

```python
# Select the first 3 characters from the affections key (general affection)
egresos['DIAG_INI'] = egresos['DIAG_INI'].str[:3]
egresos['AFECPRIN'] = egresos['AFECPRIN'].str[:3]
egresos['CAUSAEXT'] = egresos['CAUSAEXT'].str[:3]
```

```python
egresos.head()
```

```python
# Retain only the required columns
claves = claves[['CATALOG_KEY','NOMBRE']]
claves.rename(columns={'CATALOG_KEY':'CLAVE'}, inplace = True)
claves.head()
```

```python
#Check duplicates
claves[claves['CLAVE'].duplicated()]
```

```python
# Drop duplicates
claves.drop_duplicates(subset = ['CLAVE'], keep = 'last', inplace = True)
```

```python
# Merge the key table with discharges table for initial diagnosis column
egresos.rename(columns={'DIAG_INI':'CLAVE'}, inplace = True)
egresos = egresos.merge(claves, how = 'left', on = 'CLAVE')
egresos.rename(columns={'NOMBRE':'DIAG_INI'}, inplace = True)
egresos.drop(['CLAVE'], axis = 1, inplace = True)
egresos.head()
```

```python
# Merge the key table with discharges table for principal affection column.
egresos.rename(columns={'AFECPRIN':'CLAVE'}, inplace = True)
egresos = egresos.merge(claves, how = 'left', on = 'CLAVE')
egresos.rename(columns={'NOMBRE':'AFECPRIN'}, inplace = True)
egresos.drop(['CLAVE'], axis = 1, inplace = True)
egresos.head()
```

```python
# Merge the key table with discharges table for external cause column.
egresos.rename(columns={'CAUSAEXT':'CLAVE'}, inplace = True)
egresos = egresos.merge(claves, how = 'left', on = 'CLAVE')
egresos.rename(columns={'NOMBRE':'CAUSAEXT'}, inplace = True)
egresos.drop(['CLAVE'], axis = 1, inplace = True)
egresos.head()
```

```python
# Reorder columns
v_keep = ['ID', 'CLUES', 'INGRE', 'EGRESO', 'DIAS_ESTA', 'CVEEDAD','EDAD', 'NACIOEN', 'SEXO', 'PESO', 'TALLA', 'DERHAB',
         'MUNIC', 'LOC', 'TIPSERV', 'PROCED', 'DIAG_INI', 'AFECPRIN', 'VEZ', 'CAUSAEXT', 'INFEC', 'MOTEGRE']
egresos = egresos[v_keep]
egresos.head()
```

```python
#Change CVEEDAD keys to their description
egresos['CVEEDAD'].replace({0:'HORAS', 1: 'DIAS', 2:'MESES',3:'AÑOS', 9:'NE'}, inplace = True)
```

```python
#Change NACIOEN keys to their description
egresos['NACIOEN'].replace({1:'SI',2:'NO', 8:'NA', 999:'NE'}, inplace = True)
```

```python
# Change SEXO keys to their description
egresos['SEXO'].replace({1:'H', 2:'M', 9:'NE', 999:'NE'}, inplace = True)
```

```python
# Replace letters to numbers
egresos['DERHAB'].replace({'G':12}, inplace = True)
```

```python
# Change DERHAB data type
egresos['DERHAB'] = egresos['DERHAB'].astype(float)
```

```python
# Change DERHAB keys to their description
egresos['DERHAB'].replace({0:'NINGUNA', 1: 'IMSS', 2:'ISSSTE', 3:'PEMEX', 4:'SEDENA', 5:'SEMAR',
                          6:'GOBIERNO ESTATAL', 7:'SEGURO PRIVADO', 8: 'SEGURO POPULAR', 9:'NE', 10:'OTRO',
                          11:'INSABI', 12:'GRATUIDAD'}, inplace = True)
```

```python
#Change MUNIC keys to their description
egresos['MUNIC'].replace({1:'AGUASCALIENTES', 2:'ASIENTOS', 3:'CALVILLO', 4:'COSIO', 5:'JESUS MARIA',
                         6:'PABELLON DE ARTEAGA', 7:'RINCON DE ROMOS', 8:'SAN JOSE DE GRACIA',
                         9:'TEPEZALA', 10:'EL LLANO',11:'SAN FRANCISCO DE LOS ROMO', 999: 'NE'}, inplace = True)
```

```python
#Change TIPSERV keys to their description
egresos['TIPSERV'].replace({1:'NORMAL', 2:'CORTA ESTANCIA', 202:np.nan}, inplace = True)
```

```python
#Change PROCED keys to their description
egresos['PROCED'].replace({1:'CONSULTA EXTERNA', 2:'URGENCIAS', 3:'REFERIDO', 4:'OTRO',
                          6:'CUNERO PATOLOGICO', 9:'NE'}, inplace = True)
```

```python
#Replce letters with numbers
egresos['VEZ'].replace({'V486':9}, inplace = True)
#CChange data type to float
egresos['VEZ'] = egresos['VEZ'].astype(float)
# Change VEZ keys to their description
egresos['VEZ'].replace({1:'PRIMERA', 2:'SUBSECUENTE', 9:'NE'}, inplace = True)
```

```python
#Change INFEC keys to their description
egresos['INFEC'].replace({1:'SI', 2:'NO', 9:'NE'}, inplace = True)
```

```python
#Change CVEEDAD keys to their description
egresos['MOTEGRE'].replace({1:'CURACION', 2:'MEJORIA', 3:'VOLUNTARIO', 4:'TRASLADO', 5:'DEFUNCION',
                           6:'FUGA', 7:'OTRO', 9:'NE'}, inplace = True)
```
